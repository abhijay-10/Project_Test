```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text_data = [
    "NLP is amazing! Learning models and patterns.",
    "Learning NLP is fun and powerful.",
    "Deep learning improves NLP applications.",
    "Text data requires cleaning and preprocessing.",
    "Natural Language Processing is widely used in AI.",
    "AI models learn from large amounts of text data.",
    "Preprocessing text is an important step in NLP.",
    "Machine learning and deep learning are related fields."
]
```

```python
def preprocess(text):
    doc = nlp(text.lower())

    tokens = []
    for token in doc:
        if not token.is_punct and not token.is_stop:  # remove punctuation & stopwords
            tokens.append(token.lemma_)  # lemmatization (better than stemming)

    return " ".join(tokens)
```

```python
cleaned_text = [preprocess(sentence) for sentence in text_data]
```

```python
print("Cleaned Text:\n")
for line in cleaned_text:
    print(line)
```

```python
vect = CountVectorizer()
bow_matrix = vect.fit_transform(cleaned_text)

print("\nBow Matrix Shape:", bow_matrix.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(cleaned_text)

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

BoW vs TF-IDF:

BoW simply counts word occurrences without considering importance.

TF-IDF weighs words based on how unique they are across documents.

TF-IDF reduces noise from common words and gives better feature representation.
