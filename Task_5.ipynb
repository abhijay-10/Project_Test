```python
# pip install transformers huggingface_hub torch

from huggingface_hub import login
from transformers import pipeline
import os
from collections import Counter
```

```python
hf_token = os.getenv("abhi")  # created token in colab and used it

login(token=hf_token)
```

```python
classifier = pipeline("sentiment-analysis")
# studied hugging face sentiment documentation--- Also roberta can be used here
# default model: distilbert-base-uncased-finetuned-sst-2-english
```

```python
text = """
Artificial Intelligence is transforming industries across the world by enabling machines to perform tasks
that typically require human intelligence. From healthcare to finance, AI systems are improving efficiency
and decision-making. However, challenges such as data privacy, bias, and ethical concerns remain significant.
Organizations are increasingly investing in AI research to overcome these limitations and unlock its full potential.
""" * 5
```

```python

max_chunk_length = 400  # idhar length of words specify kare hain

def split_text(text, max_len):
    words = text.split()
    chunks = []
    for i in range(0, len(words), max_len):
        chunk = " ".join(words[i:i+max_len])
        chunks.append(chunk)
    return chunks

chunks = split_text(text, max_chunk_length)
```

```python
from collections import Counter

labels = []
scores = []

for chunk in chunks:
    result = classifier(chunk)[0]

    labels.append(result['label'])
    scores.append(result['score'])


final_prediction = Counter(labels).most_common(1)[0][0]


avg_confidence = sum(scores) / len(scores)


print("Chunk Predictions:", labels)
print("Final Sentiment:", final_prediction)
print("Average Confidence:", round(avg_confidence, 3))
```

Why is it beneficial to use pretrained models from Hugging Face instead of training
models from scratch? (2–3 lines)

Pretrained models are trained on massive datasets, saving huge time and compute.

They provide high accuracy with minimal fine-tuning.

This allows developers to build powerful NLP systems quickly and efficiently.
